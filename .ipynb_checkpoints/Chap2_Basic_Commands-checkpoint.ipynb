### Vectors, Data, Matrices and Subsetting


```python
x = np.array([1,3,2,5]) # create np vector
x
```

```python
x = np.array([1,6,2])
y = np.array([1,4,3])
```

```python
len(x),len(y) # length of each np array
```

```python
x + y 
```

```python
# create matrix
np.matrix([x,y])
```

```python
np.matrix(np.transpose([x,y]))
```

```python
m = np.matrix(np.arange(1,13)).reshape(4,3)
m
```

```python
np.sqrt(m)
```

```python
# generate random number
n = np.matrix(np.random.normal(0,1,10)).reshape(5,2)
n
```

```python
# subset matrix
n[:2,:2]
```

```python
# generate random number
x = np.random.normal(0,1,50)
y = x + np.random.normal(50,0.1,50)
print np.corrcoef(x,y)
print np.corrcoef(x,y)[1,0]
```

### Graphics

```python
x = np.random.normal(0,1,100)
y = np.random.normal(0,1,100)

plt.plot(x,y,"ok")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("plot of x and y")
```

```python
plt.figure(figsize = (12,5))

plt.subplot(121)    # parameter indicates 1 rows, 2 col, first figure
plt.scatter(x, y)
plt.title("figure 1")

plt.subplot(122)
plt.hist(y)
plt.title("figure 2")

plt.show()
```
